```python
def chatbot_response(user_input):
    responses = {
        "hello": "Hi there!",
        "how are you": "I'm just a bot, but I'm doing fine!",
        "what is your name": "I don't have a name, I'm just a chatbot."
    }
    return responses.get(user_input.lower(), "Sorry, I don't understand that.")

# Simulate interaction
#print(chatbot_response("Hello"))
#print(chatbot_response("how are you"))
print(chatbot_response("what is your name"))
```
